```python
from __future__ import division

import pandas as pd
import numpy as np
from sklearn.feature_extraction import DictVectorizer

%matplotlib inline
```

```python
# Build dataframes
df_train = pd.DataFrame.from_csv('train.csv', index_col=None)
df_test = pd.DataFrame.from_csv('test.csv', index_col=None)
df_prod_desc = pd.DataFrame.from_csv('product_descriptions.csv', index_col=None)
df_attrib = pd.DataFrame.from_csv('attributes.csv', index_col=None)
df_brand = df_attrib[df_attrib.name == "MFG Brand Name"][["product_uid", "value"]].rename(columns={"value": "brand"})

df = pd.merge(df_train, df_prod_desc, how='left', on='product_uid')
df = pd.merge(df, df_brand, how='left', on='product_uid')
```

```python
df.head(30)
```

```python
# Select the relevant columns (product_title, search_term, and relevance) from train df and convert to a list of dictionaries
train_dict = df.ix[:, 'product_title':'relevance'].T.to_dict().values()

# Instantiate the vectorizer
vectorizer = DictVectorizer()

# Vectorize the train data
train_matrix = vectorizer.fit_transform(train_dict).toarray()
```
